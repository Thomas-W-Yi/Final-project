# 1. INTRODUCTION

### 1.1 BACKGROUND

The Greater Toronto Area (GTA) is the most populous metropolitan area in Canada. It consists of the City of Toronto and 4 surrounding regional municipalities: Durham, Halton, Peel, and York. There are 25 incorporated municipalities in the GTA area, and is recognized as one of the most multicultural and cosmopolitan cities in the world. The metropolitan area continues to grow and attract immigrants from all over the world.
